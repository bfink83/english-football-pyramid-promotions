```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://www.skysports.com/premier-league-table'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)
```

```python
df = pd.read_html(r.text)[0]

df = df[['Team', 'Pts']]
```

```python
i = np.arange(1,21,1)
df = df.set_index(i)

df
```

```python
print("Champions League Qualifiers: ")
print(df['Team'][1], '(Champion)')
print(df['Team'][2])
print(df['Team'][3])
print(df['Team'][4])
print()
print("Europa League Qualifiers: ")
print(df['Team'][5])
print(df['Team'][6])
print()
print("Europa Conference League Qualifier: ")
print(df['Team'][7])
```
