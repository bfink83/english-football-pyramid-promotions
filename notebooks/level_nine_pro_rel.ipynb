```python
leagues = ['Combined Counties Football League North', 'Combined Counties Football League South', 
           'Eastern Counties Football League', 'Essex Senior Football League', 'Hellenic Football League',
           'Midland Football League', 'North West Counties Football League', 'Northern Counties East Football League',
           'Northern Football League', 'Southern Combination Football League', 'Southern Counties East Football League',
           'Spartan South Midlands Football League', 'United Counties League North', 'United Counties League South',
           'Wessex Football League', 'Western Football League']
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

league_ids = ['299/', '25/', '22/', '26/', '19/', '214/', '13/', '14/', '15/', '18/', '23/', '24/', '300/', '21/', '17/', '16/']
url = 'https://www.nonleaguematters.co.uk/divisions/'

# urls = ['https://www.nonleaguematters.co.uk/divisions/299/',
#         'https://www.nonleaguematters.co.uk/divisions/25/',
#         'https://www.nonleaguematters.co.uk/divisions/22/',
#         'https://www.nonleaguematters.co.uk/divisions/26/',
#         'https://www.nonleaguematters.co.uk/divisions/19/',
#         'https://www.nonleaguematters.co.uk/divisions/214/',
#         'https://www.nonleaguematters.co.uk/divisions/13/',
#         'https://www.nonleaguematters.co.uk/divisions/14/',
#         'https://www.nonleaguematters.co.uk/divisions/15/',
#         'https://www.nonleaguematters.co.uk/divisions/18/',
#         'https://www.nonleaguematters.co.uk/divisions/23/',
#         'https://www.nonleaguematters.co.uk/divisions/24/',
#         'https://www.nonleaguematters.co.uk/divisions/300/',
#         'https://www.nonleaguematters.co.uk/divisions/21/',
#         'https://www.nonleaguematters.co.uk/divisions/17/',
#         'https://www.nonleaguematters.co.uk/divisions/16/',
#        ]


regex = ['Home', 'Away', 'W', 'D', 'L', 'F', 'A', 'GD']

newTables = []
```

```python
for lid in league_ids:
    r = requests.get(url + lid, headers=header)
    df = pd.read_html(r.text)[1]
    df.columns = df.columns.to_flat_index()
    for r in regex:
        df = df[df.columns.drop(list(df.filter(regex=r)))]
    df = df.set_axis(['Pos', 'Team', 'Pld', 'Pts'], axis=1, inplace=False)
    df = df.drop(['Pos'], axis=1)
    df = df[df['Pld'].apply(lambda x: str(x).isdigit())]
    newTables.append(df)
```

```python
## NEWER VERSION ##

def getSecondPPG(tables):
    secondPlaceTeams = []
    for df in tables:
        vals = df.iloc[:,1:]
        gamesLeft = []
        numTeams = len(df)

        for i in range(0, numTeams):
            gamesLeft.append((numTeams * 2) - int(vals.iloc[i]['Pld']) - 2)

        df['Rem'] = gamesLeft
        df['Pts'] = vals['Pts']

        ppg = np.round((vals['Pts'].div(vals['Pld']).replace(np.inf, 0.00)), 3)
            
        df['PPG'] = ppg
        df = df.sort_values('PPG', ascending=False)
        
        second = df.iloc[1,:].values
        secondPlaceTeams.append(second)

    return secondPlaceTeams
```

```python
secondPlace = getSecondPPG(newTables)
```

```python
i = np.arange(1,17,1)
cols = ['Team', 'Pld', 'Pts', 'Rem', 'PPG']
DF = pd.DataFrame(columns=cols, data=secondPlace)
DF['League'] = leagues
DF = DF.sort_values('PPG', ascending=False)
DF = DF.reset_index(drop=True)
DF = DF.set_index(i)
DF.insert(1, 'League', DF.pop('League'))
```

```python
new_vals = DF.iloc[:, 2:5].values

totMatches = new_vals[:,0].astype(int) + new_vals[:,2].astype(int)
posPoints = new_vals[:,1].astype(int) + (new_vals[:,2].astype(int) * 3)
maxPPG = np.round(posPoints / totMatches, 3)

DF['max PPG'] = maxPPG

minPPG = np.round(new_vals[:,1].astype(int) / totMatches, 3)

DF['min PPG'] = minPPG

cutoff = np.sort(maxPPG)[5]
cutoff2 = np.sort(minPPG)[6]
```

```python
# Func to draw line segment
def newline(p1, p2, color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]])
    ax.add_line(l)
    return l

# Figure and Axes
fig, ax = plt.subplots(1,1,figsize=(10,7), dpi= 80)

# Vertical Lines
ax.vlines(x=cutoff, ymin=0, ymax=16, color='green', alpha=1, linewidth=1, linestyles='dotted')
ax.vlines(x=cutoff2, ymin=0, ymax=16, color='red', alpha=1, linewidth=1, linestyles='dotted')

# Points
ax.scatter(y=DF.index, x=DF['min PPG'], s=50, alpha=0.7)
ax.scatter(y=DF.index, x=DF['max PPG'], s=50, alpha=0.7)
ax.scatter(y=DF.index, x=DF['PPG'], s=50, alpha=0.7)

# Line Segments
for i, p1, p2 in zip(DF.index, DF['min PPG'], DF['max PPG']):
    newline([p1, i], [p2, i])
    
# ax.scatter(y=DF.index, x=DF['Pts'])

# Decoration
ax.set_title("Second Place Teams: Min PPG -> Max PPG", fontdict={'size':22})
ax.set_yticks(np.arange(1,17,1))
ax.set_yticklabels(DF['Team'])
# ax.set_xticks([cutoff])
# ax.set_xticklabels(["Safety Line"])
ax.invert_yaxis()
txt="Bottom six second place teams play inter-step playoff game for promotion"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
```
